### Combining multiple datasets

```python
rev2=pd.Series(data=[6210, 7410], index=["Bali", "Singapore"])
pop2=pd.Series(data=[2015, 581], index=["Bali", "Singapore"])
df_city2=pd.DataFrame({"revenue":rev2,"population": pop2})
print(df_city2)
```

```python
df_city_all=pd.concat([df_city,df_city2])
```

```python
df_city_country=pd.DataFrame({"country":["Holland", "Indonesia", "Japan","Indonesia"], 
                             "capital":[1,1,1,0]},
                              index=["Amsterdam", "Jakarta", "Tokyo","Bali"])
df_city_country
```

```python
df_all=pd.concat([df_city_all,df_city_country],axis=1, sort=False) # merging based on the column
df_all
```

```python
pd.concat([df_city_all,df_city_country],axis=1, sort=True) # merging based on the column
```

```python
pd.concat([df_city_all,df_city_country],axis=1, join="inner") # merging based on the column AND don't contain NaN value(s)
```

```python
# merging based on other column (NOT based on index)
df_country=pd.DataFrame({"pop_country":[50000,125000]}, index=["Holland","Indonesia"])
df_country
```

```python
pd.merge(df_all, df_country, left_on="country",right_index=True)
```

#### Practice
- find BOS vs LAL game between 2000-2010
- count the avg pts and opp avg points /year
- plot

```python
df_nba
```

```python
df_bl=df_nba[(df_nba["team_id"]=="BOS") & (df_nba["opp_id"]=="LAL")] 
df_lal=df_nba[(df_nba["team_id"]=="LAL") & (df_nba["opp_id"]=="BOS")]
df_lal
```

```python
df_bl
```

```python
# get the game(s) of BOS vs LAL
df_bos_lal=df_nba[(df_nba["team_id"]=="BOS") & (df_nba["opp_id"]=="LAL") & (df_nba["_iscopy"]==0) & (df_nba["yr"]<=2010) & (df_nba["yr"]>=2000)]
df_bos_lal
```

```python
# calculate the average points of each team
df_bos_avg=df_bos_lal.groupby(["yr"])["pts"].mean()
df_lal_avg=df_bos_lal.groupby(["yr"])["opp_pts"].mean()
```

```python
%matplotlib inline
df_bos_avg.plot()
df_lal_avg.plot()
plt.title("Average Points of BOS and LAL between 2000-2010")
plt.xlabel("Year")
plt.ylabel("Average Points")
leg=plt.legend()
leg.get_texts()[0].set_text('BOS')
leg.get_texts()[1].set_text('LAL')
```
