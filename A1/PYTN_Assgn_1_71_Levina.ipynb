```python
# rename the column of the coord file
borough_coord.rename(columns={'Borough': 'borough'},inplace=True)
borough_coord.head()
```

```python
# combining the coordinates with the total crime count dataframe for visualisation on a map purpose
df_map=borough_coord.merge(df_bor, on='borough', how='inner')
df_map.rename(columns={'value': 'crime'},inplace=True)

df_map.head()
```

```python
# calculating the total crime for further use in Folium map
total_crime=df_map['crime'].sum()
total_crime
```

```python
# reading the json file for London boroughs
# source: https://skgrange.github.io/data.html
geojson=r'london_boroughs.json'
```

```python
m=folium.Map(location=[51.526172028215704, -0.1290909504161549],tiles='Stamen Toner',\
           zoom_start=10.5)

#markerCluster=MarkerCluster().add_to(m)

for i, row in df_map.iterrows():
    lat=df_map.at[i, 'NS']
    lng=df_map.at[i, 'EW']
    
    pop="Cases in "+str(df_map.at[i, 'borough'])+": "+str(df_map.at[i, 'crime'])
    
    folium.CircleMarker(location=[lat,lng], popup=pop, radius=int(df_map['crime'].loc[i]/total_crime*1000), fill=True).add_to(m)
    #folium.Marker(location=[lat,lng], popup=pop, icon=folium.Icon(color='purple')).add_to(m)

folium.GeoJson(geojson,
                 name='geojson',
               style_function = lambda x: {
                   'color': 'blue',
                   'weight': 3,
                   "opacity":1,
                   'fillOpacity': 0,
                   'interactive':False
               }).add_to(m)

#m.add_child(markerCluster)
m.save('LondonCrime.html')
m
```

The map above shows London that has been overlaid by the json file to highlight the boundaries of each borough. The varying-size of blue circle is in relationship with the number of crimes that occurred in that specific borough. Once clicked, you will then be able to see the number of cases of that specific borough.

### WordCloud
<b>Question 11:</b>
- How is the distribution of each minor crime category in all boroughs?

```python
# checking the data
df_min_cat.head()
```

```python
# calculate the number of occurrence for each minor crime category
max_words=1000

all_data=""

for min_cat in df_min_cat.index.values:
    min_cat_name=min_cat.replace(" ","")
    # print(min_cat_name)
    
    repeat=int(df_min_cat.loc[min_cat]/float(total_crime)*max_words)
    # print(repeat)
    
    all_data=all_data+(str(min_cat_name)+" ")*repeat
    
# print(all_data)
```

```python
# create wordcloud with a white background 
wc_min_crime=WordCloud(background_color="white",
                        width=2000,
                        height=1000,
                        collocations=False) 

wc_min_crime.generate(all_data) 
fig=plt.figure()
fig.set_figwidth(12) # set width
fig.set_figheight(22) # set height
plt.imshow(wc_min_crime);
plt.axis("off");
```

Based on the figure above, Other Theft seems to be the #1 reason on the high crime rate in London, followed by Theft from Motor Vehicle and Burglary in a Dwelling. On the other hand, the least amount of crimes that happened are no longer visible on this image.
